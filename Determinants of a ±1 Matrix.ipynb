Display and define the "Edelman" matrix
which has entries ±1.

The determinant in exact arithmetic is even as it is a sum of an even number, factorial(27), terms all of which are ±1.

```julia
# For compactness create a 27² =  729  string of plusses and minuses
EdelmanData = "++---+-+---+-+--+++-+---++++-+-++-+++-++++-++--+-+-+--+--+++-+-------+-----+-++++++---+--++---++-++-+-++++-+--+--+-+++++-+-+++++--++++--++--+-++-+---+-------+------+---+-----++++++++-++-++++-++-+-------++--+----+--+++-+-+-+++--+-----+-+---++-+-------+-++----++-----+--+-+-+--++-+++---+-++-+++-+--+-+--++--++--+++---+++-+-+++--+++-++++---+---+++++--++-+-+-++-++++--+--+--+--++--+--++--+-----++---+-+-----+++-+---++++-++----++-+++-+-+-----++-+-+----++++--++-+-+++-+-++++++++++-++----+-+++--++--++-++---++----++++-------+--++-++-++-+-----+-+-+----+++++--+++++--+--++--+---++++-+-++-+-++++-+++++-++-++--++-++++-+--+++++-++++-+---++--++++--++-+++-+-++-+-++-+--+++-++-+-+-+----+-++-++-++++-----+--+++--+-++---+--------+----+-+--+-+-+-+"

# print row by row for human viewing
for i=0:26
    println(EdelmanData[27*i+(1:27) ])
end

# Create integer Matrix
Edelman = ones(Int,27,27)
for i=1:27, j=1:27
   if EdelmanData[27*(i-1)+j].=='-'  
        Edelman[i,j] = -1
    end
end
Edelman
```

```julia
# Use Julia's built in det which computes LU and takes ±product of the pivots
@printf("%0.2f",det(Edelman))
```

```julia
# Use BigInt's -- pushes the rounding to the very end
@printf("%0.90f",det(BigInt.(Edelman)))
```

```julia
# Use Python's symbolic package from Julia
# Pkg.add("SymPy")
using SymPy
det(Sym.(Edelman))
```

```julia
# use svd
@printf("%0.2f",prod(svdvals(Edelman)))
```

```julia
# use svd in backwards order
@printf("%0.2f",prod(svdvals(Edelman)[end:-1:1]))
```

```julia
# use pivots from lufact
@printf("%0.2f",-prod(diag(lufact(Edelman)[:U])))
```

```julia
# use backward pivots from lufact
@printf("%0.2f",-prod(diag(lu(Edelman)[2])[end:-1:1]))
```

```julia
s = svdvals(Edelman)
trials = 10000000
dets = [prod(s[randperm(27)])-839466457497600 for i=1:trials]
minimum(dets),mean(dets), maximum(dets)
```
